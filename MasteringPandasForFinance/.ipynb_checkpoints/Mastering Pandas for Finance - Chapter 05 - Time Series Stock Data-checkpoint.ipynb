```python
# Get a range of data 
stock_source = stock_source

def get_all_prices(tickers, start, end):
    
    def get_ticker_data(ticker):
        return data.DataReader(ticker, stock_source, start, end)
   
    datas = map(get_ticker_data, tickers)
    return pd.concat(datas, keys=tickers, names=['Ticker','Date'])

    '''
    results = []
    for ticker in tickers:
        results.append(data(tickers))
    
    return results
    '''
    
start = datetime.date(2012, 1, 1)
end = datetime.date(2014, 12, 31)
tickers = ['AAPL', 'MSFT', 'GE', 'IBM', 'LUV', 'DAL', 'UAL', 'PEP', 'KO']
# AAL = American Airlines. # KO = Coca-Cola
all_data = get_all_prices(tickers, start, end)
```

### Fundamental Financial Calculations





```python
# Get closing prices indexed by dates
# return = (p_t / p_t-1) - 1
daily_close_px = all_data.unstack(level=['Ticker'])['Close']
daily_close_px[:50]
```

```python
# Calculate daily return for a single price on a single day
LUV_p_t0 = daily_close_px.iloc[0]['LUV']
LUV_p_t1 = daily_close_px.iloc[1]['LUV']
r_t1 = ( LUV_p_t1 / LUV_p_t0 ) - 1   # returns
LUV_p_t0, LUV_p_t1, r_t1
```

```python
# METHOD 1 - Pandas + An Array - Calculate daily return for the entire set of stocks across all days
# Cumbersome because requires selecting from an array and dealing with both DF's and Arrays

# Value in all days + 1 / value in all days AS AN ARRAY
# Making it an array prevents Panda's otherwise very useful data alignment,
# which we actually don't want here.  
# Note: If we removed .values, we'd just get 1.0 in all result cells due to date alignment
daily_close_px.iloc[1:] / daily_close_px.iloc[:-1].values - 1
```

```python
# METHOD 2 - Pandas Shifting - Calculate daily return for the entire set of stocks across all days
# Elegant because uses Pandas's data alignment to our advantage by shifting the data forward in dates

daily_pct_change = (daily_close_px / daily_close_px.shift(1) - 1).fillna(0)
daily_pct_change.iloc[:5, :]
```

```python
# Cumulative Daily Returns
# i_t = (1 + r_t) * i_t-1 * i_0.  
        # i_0 = 1 definitionally.  If initial investment is higher, just scale i_0
# Use Pandas's broadcast of addition 
cum_daily_return = (1+ daily_pct_change).cumprod()
cum_daily_return
```

```python
# Plot Ticker value
cum_daily_return.plot(figsize=(12,8))
plt.legend(loc=2)
```

```python
msftC = msft['2012']['Close']
msftC[:5]
```

```python
# Rolling Average Plot example
memory = np.array([5,10, 20])
print memory
plt.figure(figsize=(12,8))
msftC.plot(label = 'Daily Closing Price', legend=True)
for i in range(len(memory)):
    msftC.rolling(memory[i]).mean().plot(figsize=(12,8), ylim=[0,max(msftC)], label="Rolling %s" % memory[i] + " day mean", legend=True)
```

```python
# User-defined rolling window function

# Use a lambda because the pd.rolling_apply() function will decompose the passed list
# into a list that is t objects long, where t is our rolling lag
mean_abs_dev = lambda x: np.fabs(x - x.mean()).mean()

t = 5
#pd.rolling_apply(msftC, t, mean_abs_dev).plot(figsize=(12,8))
msftC.rolling(center=False, window=t).apply(func=mean_abs_dev).plot(figsize=(12,8))

# Notice the difference here -- with this function we are calculating only one value.
# The Pandas rolling_apply function does the hard work for us of breaking the list into its rolling parts
def mean_abs_dev_now(x):
    return np.fabs(x - x.mean()).mean()

print msftC[:5].agg(mean_abs_dev_now)
print mean_abs_dev_now(msftC[:5])
```

```python
# Expanding Mean -- recalculates the mean at each subsequent time slot
msftC.plot(figsize=(12,8))
msftC.expanding(min_periods=1).mean().plot()
```

```python
# Volatility Calculation - Volatility defined as rolling window standard deviation * sqrt of period window

min_periods = 75
vol = daily_pct_change.rolling(window = min_periods, center=False).std() * np.sqrt(min_periods)
vol.plot(figsize=(10,8))
```

```python
# Volatility Calculation - Volatility defined as rolling window standard deviation 

min_periods = 75
vol = daily_pct_change.rolling(window = min_periods, center=False).std()
vol.plot(figsize=(10,8))
```

```python
# Correlation of stocks over time
min_periods = 252
rolling_corr_msft_aapl = daily_pct_change['AAPL'].rolling(window = min_periods, center = False).corr(daily_pct_change['MSFT']).dropna()
rolling_corr_msft_aapl.plot(figsize=(12,8))
```

```python
# Perform OLS regression on the % change

# Define data in a Dict
ols_data = {'AAPL': daily_pct_change['AAPL'], 'MSFT': daily_pct_change['MSFT']}
# Create an OLS model object
result = sm.ols(formula="AAPL ~ MSFT", data = ols_data)

# Call the fit method on that objecty
result = result.fit()

# Print variables held in that
print result.params
result.summary()
```
