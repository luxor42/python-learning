```python
'''
Karger's algorithm:
1. Pick an edge (a pair of vertices joined) uniformly at random
2. Merge u and v into a single vertex
3. Remove self-loops (vertices with edges pointing to themselves)

Keep track of the nodes merged into each other.
'''
```

```python
def collapse_edges(cutSet, diagnostics = False):
    
    cutList = list(cutSet.copy())
    
    while(len(cutList) > 1):
        # Initialize this to blank each run of the loop        
        trivial_edge_nodes = []
        
        # Randomly choose an edge and collapse it
        random_edge_index = random.sample(np.arange(len(cutList)),1)[0]
        comparison_edge = (cutList[random_edge_index][0], cutList[random_edge_index][1])
        collapsed_nodes = str(cutList[random_edge_index][0]) + '|' + str(cutList[random_edge_index][1])
    
        if(diagnostics == True):
            print 'collapsing nodes: ' + str(collapsed_nodes)
        # Cycle through the list of edges
        # If we encounter a node in the edge list that's attached to the edge we're collapsing
        # Then replace that node with the collapsed nodes

        for idx, edge in enumerate(cutList):
            if(diagnostics == True):
                print ''
                print edge[0],'++' ,edge[1],  '-----', comparison_edge[0], '++', comparison_edge[1]  

            if edge[0] == comparison_edge[0] or edge[0] == comparison_edge[1]:
                # Output
                if(diagnostics == True):                
                    print '\t' + 'Match with ' + str(edge[0])
                    print 'original value: ', cutList[idx]
                    
                # Operation
                cutList[idx] = (collapsed_nodes, edge[1])
                
                # Output
                if(diagnostics == True):
                    print 'new value: ', cutList[idx]            

            if edge[1] == comparison_edge[0] or edge[1] == comparison_edge[1]:
                # Output
                if(diagnostics == True):
                    print '\t' + 'Match with ' + str(edge[1]) 
                    print 'original value: ', cutList[idx]  

                # Operation
                # Set it equal to cutList[idx][0] instead of edge[0] because may have updated cutList above
                cutList[idx] = (cutList[idx][0], collapsed_nodes)

                # Output
                if(diagnostics == True):
                    print 'new value: ', cutList[idx]

            # If we have created an edge that maps to itself, then track its index for deletion later
            # Note that the target edge to remove will always be a trivial edge node
            if cutList[idx][0] == cutList[idx][1]:
                trivial_edge_nodes.append(idx)

        # Remove the edge we're collapsing.
        # Do the deletion afterwards so we can keep using the same index value

        # Output
        if(diagnostics == True):
            print 'FOUND TRIVIAL EDGE NODES.  REMOVING FROM SET'
            
        # Operation
        for i in sorted(trivial_edge_nodes, reverse=True):
            # Output
            if(diagnostics == True):
                print 'Removing values at index: ', i, '... value: ', cutList[i]   
                
            # Operation
            del cutList[i]
            
            # Output
            if(diagnostics == True):
                print cutList

        # Remove duplicate values
        cutList = list(set(cutList))
        
        # Once we're down to 2 nodes we need to make sure they're not just inverses of each other
        if(len(cutList) == 2):
            if cutList[0][0] == cutList[1][1] and cutList[0][1] == cutList[1][0]:
                del cutList[0]
                   
    return cutList

def parse_sets(cutList, diagnostics = False):
    # Parse the left and right sets
    left_set = [int(x) for x in str(cutList[0][0]).split('|')]
    right_set = [int(x) for x in str(cutList[0][1]).split('|')]
    if(diagnostics == True): 
        print 'final set: ' + str(cutList)
        print 'Left set: ' + str(left_set)
        print 'Right set: ' + str(right_set)

    # Cycle through the original set.
    # If first node is in the left set, then check if its pair is in the right set.  If so, add 1 to counter.
    # If first node is in the right set, then ...........................left set.   If so, add 1 to counter.

    return (left_set, right_set)
   
def count_cuts(left_set, right_set, initialSet, diagnostics = False):
    
    cut_counter = 0
        
    if(diagnostics == True):
        print 'Left  Cut: ', left_set, ' Right Right: ', right_set
        print initialSet

    for edges in initialSet:
        if edges[0] in left_set:
            #print str(edges[0]) + ' found in left set'
            if edges[1] in right_set:
                cut_counter = cut_counter + 1
                if(diagnostics == True):
                    print str(edges[0]) + ' found in left set. ' + str(edges[1]) + ' found in right set. COUNT IT.'

        if edges[0] in right_set:
            #print str(edges[0]) + ' found in right set'
            if edges[1] in left_set:
                if(diagnostics == True):                
                    print str(edges[0]) + ' found in right set. ' + str(edges[1]) + ' found in left set. COUNT IT.'            
                cut_counter = cut_counter + 1

    if(diagnostics == True):
        print 'Total Cuts : ' + str(cut_counter)
                   
    return cut_counter
    

def run_cut(initialSet, diagnostics = False):

    # Collapse the edges
    cutList = collapse_edges(initialSet, diagnostics)
    
    if(diagnostics == True):
        print 'Collapsed List: ', cutList
    
    # Parse the list into its left and right sides.
    result = parse_sets(cutList, diagnostics)
    left_set = result[0]
    right_set = result[1]
    
    if(diagnostics == True):
        print 'Left  Set: ', left_set, ' Right Set: ', right_set
    
    # count cuts
    return {
        'left_set': left_set,
        'right_set': right_set,
        'count': count_cuts(left_set, right_set, initialSet, diagnostics)}
     

##################################
random.seed(1)
cut_count = []
elapsed = []

diagnostic_flag = False
runs_left = 10000
runs_done = 0

start_time = time.time()

cut_count = []
left_set = []
right_set = []
minimum_cut_index = []
minimum_cut_value = False

print 'Making ' + str(runs_left) + ' runs in total'
while(runs_done < runs_left):

    result = run_cut(startingSet, diagnostic_flag)
    
    cut_count.append(result['count'])
    left_set.append(result['left_set'])
    right_set.append(result['right_set'])    
    
    if(diagnostic_flag == True):
        print 'cut count is ', result['count']
        print 'current minimum cut value is ', minimum_cut_value, ' at index ' , minimum_cut_index
        print 'this cut location is ', runs_done
        print ''

    if(result['count'] <= minimum_cut_value or minimum_cut_value == False):
        minimum_cut_value = result['count']
        # This array is wrong; is tracking minimum cut each time a new minimum is found
        # Should refactor so it tracks the index for each minimum cut found
        minimum_cut_index.append(runs_done)
    
    runs_done = runs_done + 1
    
    elapsed.append(time.time() - start_time)
    
    if runs_done % 100 == 0:
        print 'With ' + str(runs_done) + ' runs done, time elapsed = ' + str(time.time() - start_time) + ' seconds'

print cut_count   
print 'Mininum cut: ' , min(cut_count)
print 'Occurs at index/es: ', minimum_cut_index
print 'Left set first index: ', left_set[min(minimum_cut_index)]
print 'Right set first index: ', right_set[min(minimum_cut_index)]
```
